```python
DATA_DIR = Path("..") / "TakeHomeProject"
columns_path = DATA_DIR / "census-bureau.columns"
data_path = DATA_DIR / "census-bureau.data"

columns = columns_path.read_text(encoding="utf-8").strip().splitlines()

df = pd.read_csv(
    data_path,
    names=columns,
    na_values=["?", " ?", "None", ""],
    skipinitialspace=True,
    engine="c",
)

df["label"] = (
    df["label"]
    .str.strip()
    .str.replace(".", "", regex=False)
    .replace({"- 50000": 0, "50000+": 1})
    .astype(int)
)

# df.head(3)
```

```python
corr = df[numeric_cols].corr(method="pearson")
mask = np.triu(np.ones_like(corr, dtype=bool))
plt.figure(figsize=(12, 10))
sns.heatmap(
    corr,
    mask=mask,
    cmap="coolwarm",
    center=0,
    square=True,
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    cbar_kws={"shrink": 0.8},
)
plt.title("Numeric Feature Correlation (Pearson)")
plt.tight_layout()
plt.show()

corr
```

```python
encoded_cats = pd.get_dummies(df[categorical_cols], drop_first=True, dtype=float)
feature_matrix = pd.concat([df[numeric_cols], encoded_cats], axis=1)

corr_features = feature_matrix.corr()
feature_strength = corr_features.abs().sum().sort_values(ascending=False)
top_features = feature_strength.head(30).index
heatmap = corr_features.loc[top_features, top_features]
mask = np.triu(np.ones_like(heatmap, dtype=bool))

plt.figure(figsize=(14, 10))
sns.heatmap(
    heatmap,
    mask=mask,
    cmap="coolwarm",
    center=0,
    annot=False,
    linewidths=0.2,
    cbar_kws={"shrink": 0.8},
)
plt.title("Feature Correlation Heatmap (Numeric + Encoded Categorical)")
plt.tight_layout()
plt.show()

heatmap
```

```python
features = [col for col in numeric_cols if df[col].nunique(dropna=True) > 1]
positive_label = df["label"].value_counts().idxmin()
cols = 3
rows = int(np.ceil(len(features) / cols)) if features else 0
if rows == 0:
    raise ValueError("No numeric features available for plotting")
fig, axes = plt.subplots(rows, cols, figsize=(cols * 6.0, rows * 4.5), squeeze=False)
axes = axes.flatten()
feature_summaries = {}
for ax, feature in zip(axes, features):
    valid = df[[feature, "label"]].dropna()
    bins = pd.cut(valid[feature], bins=10, include_lowest=True)
    summary = (
        valid.assign(bin=bins)
        .groupby("bin", observed=False)
        .agg(
            count=("label", "size"),
            positive_rate=(
                "label",
                lambda s: (s == positive_label).mean() if len(s) else np.nan,
            ),
        )
    )
    if summary.empty:
        ax.set_visible(False)
        continue
    summary["positive_rate"] = summary["positive_rate"] * 100
    summary["positive_rate"] = summary["positive_rate"].fillna(0)
    feature_summaries[feature] = summary
    intervals = summary.index
    centers = [interval.mid for interval in intervals]
    widths = [interval.length * 0.9 for interval in intervals]
    labels = []
    for interval in intervals:
        left = interval.left
        right = interval.right
        if max(abs(left), abs(right)) >= 1000:
            labels.append(f"{left:.0f}-{right:.0f}")
        else:
            labels.append(f"{left:.1f}-{right:.1f}")
    ax.bar(
        centers,
        summary["count"],
        width=widths,
        color="steelblue",
        edgecolor="black",
        alpha=0.7,
    )
    ax.set_title(feature.replace("_", " ").title())
    ax.set_xlabel(feature)
    ax.set_ylabel("Count")
    ax.set_xticks(centers)
    ax.set_xticklabels(labels, rotation=45, ha="right", fontsize=8)
    ax.grid(axis="y", linestyle="--", alpha=0.3)
    ax2 = ax.twinx()
    ax2.plot(
        centers, summary["positive_rate"], color="crimson", marker="o", linewidth=2
    )
    ax2.set_ylabel(f"{positive_label} rate (%)")
    max_rate = summary["positive_rate"].max()
    rate_ylim = 5 if pd.isna(max_rate) else max(5, max_rate * 1.1)
    ax2.set_ylim(0, rate_ylim)
for ax in axes[len(features) :]:
    ax.set_visible(False)
fig.suptitle("Feature Distributions and Positive Rate by Decile", fontsize=16, y=0.995)
plt.tight_layout(rect=[0, 0, 1, 0.98])
fig.subplots_adjust(hspace=0.45, wspace=0.3)
plt.show()
```

```python
categorical_features = [
    col
    for col in categorical_cols
    if df[col].nunique(dropna=True) > 1 and col != "label"
]
positive_label = df["label"].value_counts().idxmin()
max_categories = 12
cols = 3
rows = int(np.ceil(len(categorical_features) / cols)) if categorical_features else 0
if rows == 0:
    raise ValueError("No categorical features available for plotting")
fig, axes = plt.subplots(rows, cols, figsize=(cols * 6.0, rows * 7.0), squeeze=False)
axes = axes.flatten()
categorical_summaries = {}
for ax, feature in zip(axes, categorical_features):
    valid = df[[feature, "label"]].dropna().copy()
    counts = valid[feature].value_counts()
    if len(counts) > max_categories:
        top = counts.nlargest(max_categories - 1).index
        valid[feature] = valid[feature].where(valid[feature].isin(top), "Other")
    summary = (
        valid.groupby(feature)
        .agg(
            count=("label", "size"),
            positive_rate=(
                "label",
                lambda s: (s == positive_label).mean() if len(s) else 0.0,
            ),
        )
        .sort_values("count", ascending=False)
    )
    summary["positive_rate"] = summary["positive_rate"] * 100
    categorical_summaries[feature] = summary
    positions = np.arange(len(summary))
    ax.bar(positions, summary["count"], color="steelblue", edgecolor="black", alpha=0.7)
    ax.set_title(feature.replace("_", " ").title())
    ax.set_xlabel(feature)
    ax.set_ylabel("Count")
    ax.set_xticks(positions)
    ax.set_xticklabels(summary.index.astype(str), rotation=45, ha="right", fontsize=8)
    ax.grid(axis="y", linestyle="--", alpha=0.3)
    ax2 = ax.twinx()
    ax2.plot(
        positions, summary["positive_rate"], color="crimson", marker="o", linewidth=2
    )
    ax2.set_ylabel(f"{positive_label} rate (%)")
    max_rate = summary["positive_rate"].max()
    rate_ylim = 5 if pd.isna(max_rate) else max(5, max_rate * 1.1)
    ax2.set_ylim(0, rate_ylim)
for ax in axes[len(categorical_features) :]:
    ax.set_visible(False)
fig.suptitle("Categorical Feature Counts and Positive Rate", fontsize=16, y=0.995)
plt.tight_layout(rect=[0, 0, 1, 0.98])
fig.subplots_adjust(hspace=0.45, wspace=0.3)
plt.show()
```

### Feature engineering

```python
import numpy as np
import pandas as pd

FEATURES = df.copy()

# Numeric ratios (guard against zero/NaN)
FEATURES["ratio_capital_gain_wage"] = np.where(
    FEATURES["wage per hour"] > 0,
    FEATURES["capital gains"] / FEATURES["wage per hour"],
    0,
)
FEATURES["ratio_dividend_income"] = np.where(
    (FEATURES["capital gains"] + FEATURES["dividends from stocks"]) > 0,
    FEATURES["dividends from stocks"]
    / (FEATURES["capital gains"] + FEATURES["dividends from stocks"]),
    0,
)
FEATURES["employment_intensity"] = np.where(
    FEATURES["age"] > 0,
    FEATURES["weeks worked in year"] / FEATURES["age"],
    0,
)

# Household composition proxies
family_lookup = {
    "Both parents, wife present": 2,
    "Both parents": 2,
    "Neither parent, other relative": 1,
    "Neither parent": 0,
}


# Winsorized / log transforms to tame skew
def log_winsor(series, lower=0.01, upper=0.99):
    clipped = series.clip(series.quantile(lower), series.quantile(upper))
    return np.log1p(clipped)


for col in ["capital gains", "capital losses", "dividends from stocks"]:
    FEATURES[f"log_{col.replace(' ', '_')}"] = log_winsor(FEATURES[col])

# Combine back with the label you need for plotting
FEATURES["label"] = df["label"]
engineered_cols = [
    "ratio_capital_gain_wage",
    "ratio_dividend_income",
    "employment_intensity",
]
FEATURES[engineered_cols + ["label"]].head()
```

```python
positive_label = FEATURES["label"].value_counts().idxmin()
cols = 3
rows = int(np.ceil(len(engineered_cols) / cols)) if engineered_cols else 0

fig, axes = plt.subplots(rows, cols, figsize=(cols * 6.0, rows * 4.5), squeeze=False)
axes = axes.flatten()
feature_summaries = {}
for ax, feature in zip(axes, engineered_cols):
    valid = FEATURES[[feature, "label"]].dropna()
    bins = pd.cut(valid[feature], bins=10, include_lowest=True)
    summary = (
        valid.assign(bin=bins)
        .groupby("bin", observed=False)
        .agg(
            count=("label", "size"),
            positive_rate=(
                "label",
                lambda s: (s == positive_label).mean() if len(s) else np.nan,
            ),
        )
    )
    if summary.empty:
        ax.set_visible(False)
        continue
    summary["positive_rate"] = summary["positive_rate"] * 100
    summary["positive_rate"] = summary["positive_rate"].fillna(0)
    feature_summaries[feature] = summary
    intervals = summary.index
    centers = [interval.mid for interval in intervals]
    widths = [interval.length * 0.9 for interval in intervals]
    labels = []
    for interval in intervals:
        left = interval.left
        right = interval.right
        if max(abs(left), abs(right)) >= 1000:
            labels.append(f"{left:.0f}-{right:.0f}")
        else:
            labels.append(f"{left:.1f}-{right:.1f}")
    ax.bar(
        centers,
        summary["count"],
        width=widths,
        color="steelblue",
        edgecolor="black",
        alpha=0.7,
    )
    ax.set_title(feature.replace("_", " ").title())
    ax.set_xlabel(feature)
    ax.set_ylabel("Count")
    ax.set_xticks(centers)
    ax.set_xticklabels(labels, rotation=45, ha="right", fontsize=8)
    ax.grid(axis="y", linestyle="--", alpha=0.3)
    ax2 = ax.twinx()
    ax2.plot(
        centers, summary["positive_rate"], color="crimson", marker="o", linewidth=2
    )
    ax2.set_ylabel(f"{positive_label} rate (%)")
    max_rate = summary["positive_rate"].max()
    rate_ylim = 5 if pd.isna(max_rate) else max(5, max_rate * 1.1)
    ax2.set_ylim(0, rate_ylim)
for ax in axes[len(engineered_cols) :]:
    ax.set_visible(False)
fig.suptitle("Feature Distributions and Positive Rate by Decile", fontsize=16, y=0.995)
plt.tight_layout(rect=[0, 0, 1, 0.98])
fig.subplots_adjust(hspace=0.45, wspace=0.3)
plt.show()
```
